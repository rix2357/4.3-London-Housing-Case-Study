To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 

To this end, you *could* apply a ***lambda function***. Your logic could work as follows:
1. look through the `Month` column
2. extract the year from each individual value in that column 
3. store that corresponding year as separate column. 

Whether you go ahead with this is up to you. Just so long as you answer our initial brief: which boroughs of London have seen the greatest house price increase, on average, over the past two decades? 

```python
# settled on using groupby function, had problems before because forgot to convert datatypes to float up in previous step
Boroughs_year = Boroughs.groupby(Boroughs.index.year).mean()
```

```python
#transpose so can add new data column calculating percentage
Boroughs_year_t = Boroughs_year.transpose()
Boroughs_year_t['% increase last 20 yr'] = Boroughs_year_t[2021]/Boroughs_year_t[2001] * 100
```

```python
TopBoroughsLast20Yr = Boroughs_year_t.sort_values('% increase last 20 yr', ascending=False)[:5]
```

```python
TopBoroughsLast20Yr['% increase last 20 yr']
```

```python
#check with graph
values = Boroughs.loc['2001':'2021',TopBoroughsLast20Yr[2021].sort_values(ascending=False).index]
plt.plot(values, label=TopBoroughsLast20Yr.index)
plt.legend()
plt.show()
```

```python
Boroughs_year_t['% increase 2018-1998'] = Boroughs_year_t[2018]/Boroughs_year_t[1998] * 100
TopBoroughs2018_1998 = Boroughs_year_t.sort_values('% increase 2018-1998', ascending=False)[:15]
TopBoroughs2018_1998['% increase 2018-1998']
```
